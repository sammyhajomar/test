```python
import altair as alt
import pandas as pd
from vega_datasets import data
import os
```

```python
#alt.data_transformers.enable('default',max_rows=5000)
alt.data_transformers.enable('json')
path = os.path.join(os.getcwd(),'opioid_dataset','Wide_Master.csv')
df_w = pd.read_csv(path,encoding = "ISO-8859-1")
df_w1 = df_w[['State','Year','Total','Income']]
#df_w['Year']= pd.to_datetime(df_w['Year'],format="%Y")
df_w1.head()
```

```python
interval = alt.selection_interval(encodings=['x','y'])

slider = alt.binding_range(min=2011, max=2017, step=1)
select_year = alt.selection_single(name="Year", fields=['Year'],
                                   bind=slider, init={'Year': 2011})

chart = alt.Chart(df_w1).mark_point().encode(
    x=alt.X('mean(Income):Q', title='Average Income ($)', scale=alt.Scale(domain=(51000,106000))),
    y=alt.Y('mean(Total):Q', title='Average Deaths Number of from Opioid Overdose', scale=alt.Scale(domain=(10,200))),
    color=alt.Color('State:N',legend=None),
    tooltip=('State:N','mean(Total):Q','mean(Income):Q'),
).add_selection(
    select_year
).transform_filter(
    select_year
).properties(
    width=850,
    height=600
)

text = alt.Chart(df_w1).mark_text(
    align= 'left',
    baseline = 'middle',
    dx= 7
).encode(
    x=alt.X('mean(Income):Q', title='Average Income ($)', scale=alt.Scale(domain=(51000,106000))),
    y=alt.Y('mean(Total):Q', title='', scale=alt.Scale(domain=(10,200))),
    text='State'
).transform_filter(
    select_year
).properties(
    width=850,
    height=600
)
chart + text
#chart | chart.encode(x='Acceleration')
```
